```python
import folium
m = folium.Map(location=[30.102261,-81.7118], zoom_start=6, tiles ="Mapbox Bright")
fg = folium.FeatureGroup(name ="Some icons")
import pandas
data = pandas.read_csv("FL_insurance_sample.csv")
lat = list(data["point_latitude"])
lon=list(data["point_longitude"])
policy = list(data["policyID"])
point = list(data["point_granularity"])
#print(type(point))
def color_change(pt):
    if pt<= 2:
       return 'blue'
    else:
       return 'green'

for lt,ln,p,pnt in zip(lat,lon,policy,point):
    #fg.add_child(folium.Marker(location=[lt,ln], popup=str(p), icon = folium.Icon(color=color_change(pnt))))
    fg.add_child(folium.CircleMarker(location =[lt,ln], radius =2, popup = str(p), fill_color = color_change(pnt),color =color_change(pnt),fill_opacity = 0.7))
fgp =folium.FeatureGroup(name ="Population")
#fg.add_child(folium.GeoJson(data=(open("world.json"),"r")))
fgp.add_child(folium.GeoJson(data=(open('world.json','r',encoding ='utf-8-sig').read()),style_function=lambda x:{'fillColor':'green'
                                                                                                               if x['properties']['POP2005']<10000000
                                                                                                               else 'orange' if 10000000 <= x['properties']['POP2005']<=50000000
                                                                                                               else 'red'}))
m.add_child(fg)
m.add_child(fgp)
m.add_child(folium.LayerControl())
m
```
