```python
import pandas
import folium
i=0
j=0
k=0
v="Un:"
m= folium.Map(location=[0,0],zoom_start=5,tiles="MapBox Bright")# reference map
fgu = folium.FeatureGroup(name="Universities-world") # another layer for universites
data = pandas.read_csv("trial2.csv")
data_n = pandas.read_csv("convertcsv.csv")
#print(data["CountryName"][i])
#print(data_n["CountryName"][j])
l_c = list(data["CountryName"])
lat = list(data["CapitalLatitude"])
#print(lat)
lon = list(data["CapitalLongitude"])
#print("lat over")
#print(lon)
for l,lt,ln in zip(l_c,lat,lon):
    while(l==data_n["CountryName"][j]):
            v = v +','+ data_n["University Name"][k]
            j=j+1
            k=k+1
    #print(v)
    #print(str(lt)+','+str(ln))
    fgu.add_child(folium.Marker(location =[lt,ln],popup=v,icon=folium.Icon(color='blue') ))
    v="Un:"
m.add_child(fgu)
m
```
