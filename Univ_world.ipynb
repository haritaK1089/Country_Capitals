```python
####################################################################################################################
# filename Univ_world
# Executing this file gives output display of world map displaying capitals of every country and Universities located in Afghanistan
# version 1
#Author: Harita.K
############################################################################################################################

import folium # import folium lib into python
import pandas #import pandas(data analysis lib) into python
m= folium.Map(location=[0,0],zoom_start=5,tiles="MapBox Bright")# reference map 
fg = folium.FeatureGroup(name="Country-capitals")# one layer for country capitals
fgu = folium.FeatureGroup(name="Universities-world") # another layer for universites
data = pandas.read_csv("country-capitals.csv")# importing data from csv file
lat = list(data["CapitalLatitude"]) #extract latitude values
#print(lat)
lon = list(data["CapitalLongitude"]) #extract longitude values
#print(lon)
for lt,ln in zip(lat,lon): #for every lat and lon create a circle marker
    fg.add_child(folium.CircleMarker(location=[float(lt),float(ln)], radius=2, color='red', fill_color='orange'))
    
m.add_child(fg)
data_u = pandas.read_csv("trial.csv")# 
#name = list(data_u["CountryName"])#country names
lat_n =list(data_u["Latitude"]) #extract latitude values
lon_n = list(data_u["Longitude"])#extract longitude values
name = list(data_u["University Name"])#university names
for lt_n,ln_n,un in zip(lat_n,lon_n,name):
    fgu.add_child(folium.Marker(location =[lt_n,ln_n],popup=un,icon=folium.Icon(color='blue') ))
m.add_child(fgu)
m.add_child(folium.LayerControl()) #Implement layer control
m.save("captl_univ.html") # save the map to captl_univ.html file
m
```
